```python
# Remove the non numeric values
df["Price"] = df["Price"].str.replace("Rs", "").str.replace(",", "").str.replace(" ","")
df["Price"] = df["Price"].str.replace("Negotiable", "")
df["Mileage"] = df["Mileage"].str.replace("km", "").str.replace(",", "").str.replace(" ","")
df["Mileage"] = df["Mileage"].str.replace("-", "")

# convert into a numeric value
df["Price"] = pd.to_numeric(df["Price"])
df["Mileage"] = pd.to_numeric(df["Mileage"])
```

# Cumalative Frequency

```python
review_freq = n_df.Review.value_counts()
```

```python
cum_df = pd.DataFrame({'Review': review_freq.index, 'Frequency': review_freq.values})
```

```python
cum_df.head()
```

```python
cum_df['Cumulative Frequency'] = cum_df['Frequency'].cumsum()
```

```python
cum_df.head()
```

# Step Chart

```python
# plt.step(cum_df['Review'], cum_df['Frequency'], 'r*', where='pre')
# plt.show()

# Create the step chart
plt.step(cum_df['Review'], cum_df['Cumulative Frequency'], where='post')

# Add axis labels and title
plt.xlabel('Review')
plt.ylabel('Cumulative Frequency')
plt.title('Cumulative Frequency of Reviews')

# Show the plot
plt.show()
```

```python
review_count_new = n_df[n_df['Condition'] == 'New']['Review'].value_counts()
review_count_new
```

```python
fr_df = pd.DataFrame({'Review': review_count_reconditioned.index, 'Count for Reconditioned Condition': review_count_reconditioned.values})
fr_df.head()
```

```python
fn_df = pd.DataFrame({'Review': review_count_new.index, 'Count for New Condition': review_count_new.values})
fn_df.head()
```

```python
import numpy as np

x = np.arange(5)
width = 0.2

plt.bar(x-0.2,fu_df['Count for Used Condition'],width, color='blue')
plt.bar(x,fr_df['Count for Reconditioned Condition'], width, color='red')
plt.bar(x+0.2,fn_df['Count for New Condition'],width, color='green')

plt.xticks(x, n_df['Review'].unique())
plt.xlabel("Review")
plt.ylabel("Counts")
plt.legend(["Used", "Reconditioned", "New"])
plt.show()
```

# Stack bar chart

```python
x = n_df['Review'].unique()
y3 = fn_df['Count for New Condition']
y2 = fr_df['Count for Reconditioned Condition']
y1 = fu_df['Count for Used Condition']

plt.bar(x, y1, color='blue')
plt.bar(x, y2, bottom=y1, color='red')
plt.bar(x, y3 ,bottom=y1+y2, color='green')

plt.xlabel("Review")
plt.ylabel("Counts")
plt.legend(["Used", "Reconditioned", "New"])
plt.show()
```

# Correlation

```python
# calculate covariance matrix
covariance = np.cov(n_df['Price'], n_df['Mileage'])
print(covariance)
```

# Pearson’s correlation

```python
from scipy.stats import pearsonr

# calculate Pearson's correlation
corr, _ = pearsonr(n_df['Price'], n_df['Mileage'])
print('Pearsons correlation: %.3f' % corr)
```

# Spearman’s Correlation

```python
from scipy.stats import spearmanr

# calculate spearman's correlation
corr, _ = spearmanr(n_df['Price'], n_df['Mileage'])
print('Spearmans correlation: %.3f' % corr)
```

```python
import seaborn as sns

corrmat = n_df.corr()
  
f, ax = plt.subplots(figsize =(9, 8))
sns.heatmap(corrmat, ax = ax, cmap ="YlGnBu", linewidths = 0.1)
```

```python
corrmat
```

```python
# Price correlation matrix
# k : number of variables for heatmap
k = 20
  
cols = corrmat.nlargest(k, 'Price')['Price'].index
  
cm = np.corrcoef(n_df[cols].values.T)
f, ax = plt.subplots(figsize =(12, 10))
  
sns.heatmap(cm, ax = ax, cmap ="YlGnBu",
            linewidths = 0.1, yticklabels = cols.values, 
                              xticklabels = cols.values)
```

```python
cg = sns.clustermap(corrmat, cmap ="YlGnBu", linewidths = 0.1);
plt.setp(cg.ax_heatmap.yaxis.get_majorticklabels(), rotation = 0)
```
